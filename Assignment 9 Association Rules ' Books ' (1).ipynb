# Apriori Algorithm

```python
frequent_itemsets=apriori(data_cleaned1, min_support=0.2, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```
